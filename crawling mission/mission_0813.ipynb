```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import re

service = Service(ChromeDriverManager().install())
options = Options()
# options.add_argument('--headless')
# options.add_argument('--disable-gpu')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-size=1920x1080')
options.add_argument('--start-maximized')
options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36')
options.add_argument('--lang=ko_KR')

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.musinsa.com/main/musinsa/ranking?gf=A&storeCode=musinsa&sectionId=200&contentsId=&categoryCode=000&ageBand=AGE_BAND_ALL"
driver.get(url)
time.sleep(3)

products = driver.find_elements(By.CSS_SELECTOR, "article.sc-qcgyze-0.frTbIc")[:50]
product_data = []

for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1.5)

product_links = []
products = driver.find_elements(By.CSS_SELECTOR, "div.sc-1t5ihy5-1.jzppnI > a")
for p in products[:50]:
    link = p.get_attribute("href")
    if link:
        product_links.append(link)
print(f"수집된 상품 링크 수: {len(product_links)}")

rows = []

for idx, link in enumerate(product_links, start=1):
    print(f"{idx}번째 수집: {link} / {product_name} / {brand} / {price} / {discount_percent} / {reviews}")
    driver.get(link)
    time.sleep(1.5)

    try:
        product_name = driver.find_element(By.CSS_SELECTOR, "span.text-title_18px_med.sc-1omefes-1.exqQRL.font-pretendard").text.strip()
    except:
        product_name = None

    try:
        brand = driver.find_element(By.CSS_SELECTOR, "span.text-body_14px_med.font-pretendard").text.strip()
    except:
        brand = None

    try:
        price = driver.find_element(By.CSS_SELECTOR, "span.text-title_18px_semi.sc-1hw5bl8-7.kXhdZT.text-black.font-pretendard").text.strip()
    except:
        price = None

    try:
        discount_percent = driver.find_element(By.CSS_SELECTOR, "span.text-title_18px_semi.sc-1hw5bl8-6.hROMjI.text-red.font-pretendard").text.strip()
    except:
        discount_percent = None

    try:
        reviews = driver.find_element(By.CSS_SELECTOR, "span.text-body_13px_reg.underline.text-gray-600.font-pretendard").text.strip()
    except:
        reviews = None

    rows.append({
        "순위": idx,
        "상품명": product_name,
        "브랜드": brand,
        "가격": price,
        "할인율": discount_percent,
        "리뷰수": reviews,
        "URL": link
    })

df = pd.DataFrame(rows)
df.to_csv("musinsa_top50.csv", index=False, encoding="utf-8-sig")
driver.quit()

print("✅ 저장 완료")
       
```
