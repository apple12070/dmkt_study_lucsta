#### 1. 데이터베이스 = DB
- 데이터베이스 => 데이터를 보관하고 저장할 수 있는 큰 공간
- 데이터베이스 역사 : 오래되었음
- 데이터를 담고 관리할 수 있는 시스템, 체계의 역사 : 오래되었음
- DBMS 역사 : 오래되었음
- 80여개 정도의 DBMS 존재
- 관계형 데이터베이스 관리 시스템 : RDBMS
- Oracle > MySQL (*원래 비등비등했으나 Oracle이 MySQL 인수하면서 더 커짐)
- SQL -> RDBMS 기반하에 저장 및 보관되어있는 데이터를 처리하기 위한 목적의 단어
- 스마트폰 -> 모바일 -> 데이터 기하급수적 증가
- 빅데이터 : 하루에도 수백.수천만건의 데이터가 쌓여지는 것을 의미
- RDBMS -> DB -> Table -> Schema
- 빅데이터라는 방대한 데이터의 입.출력건이 수시로 실행되고 있는데, 이것을 과거의 문법 및 방법으로 처리.관리한다는 것이 현실.물리적으로 부담 시작
- 시대의 흐름에 따라서 데이터를 처리할 수 있는 프로그래밍 언어가 다양해짐
- 각각의 언어가 매번 RDBMS와의 상호작용을 하게끔 처리하는 것도 현실적으로 불가능
- 더이상 SQL 문법만 가지고 RDBMS라는 데이터 관리 시스템만 의존하는 것은 구시대적인 사고방식!!!
- 그래서 등장한 것이 NoSQL ( => Not Only SQL)
- NoSQL 기반의 데이터 저리 => 전통적인 역사를 가지고 있는 기업은 아님. (신규스타트업, 유니콘기업, 창업기업 등은 SQL과 NoSQL 같이 씀)
- SQL 문법을 사용할 수 있는 Oracle, MySQL
- NoSQL 기반의 데이터를 처리할 수 있는 프로그램 => MongoDB

#### 2. NoSQL 이해
- NoSQL은 스키마 & JOIN 등의 문법 및 관리시스템이 존재하지 않음.
- 현재 작업하는 업무가 95% read, 5% write : RDBMS 100%로 사용.
- 현재 작업하는 업무에서 50% 이상 write 업무를 하고 있다 => 5:5 정도로 NoSQL + SQL

#### 3. MongoDB
- document 기반 NoSQL 데이터베이스
- MongoDB에 저장.관리되는 데이터가 BSON(Binary JSON) 기반의 document 형태로 데이터 관리
- SQL // python => API Data => json,xml
- python =>
```
document = {
    "id":"01",
    "language":"java",
    "edition":{
        "first":"1st",
        "second":"2nd",
        "third":"3rd"
    }
}
```

```
{
    "_id":ObjectId("50342khkblkj453lw23"),
    "username":"david",
    "name":{
        "first":"1st",
        "second":"2nd"
    }
}
```
- SQL : Database - Table(*Schema) - data
- NoSQL : Database - Collection - Document

#### 4. 프로그래밍 언어를 보다 효율적으로 학습하는 방법
- CRUD 입각해서 학습
- CREATE
- READ
- UPDATE
- DELETE

- show dbs : 로컬컴퓨터 내 생성된 db목록 확인
- use dbname : 특정 db로 접속하겠다는 의미
- show collections : 해당 db안에 생성되어있는 컬렉션 목록 확인
  
- db.컬렉션명.find() : 특정 db안에 컬렉션 내부에 어떤 데이터가 있는지 확인
- db.stats() : db의 상태정보를 확인
- db.dropDatabase() : db를 삭제하고자 할 때

- db.createCollection("컬렉션명") : 해당 db안에 특정 컬렉션을 생성
- db.createCollection("컬렉션명", options) :
  - capped : true => 고정된 크기를 갖고 있는 컬렉션을 만들겠다는 의미
  - autoIndexId : true => _id 필드에 대한 인덱스를 자동으로 생성하겠다
  - size : capped의 옵션 true인 경우에만 사용 => 바이트 단위
  - max : 컬렉션에 저장할 수 있는 문서의 최대 개수를 정의
- db.컬렉션명.drop() : 특정 db안에 컬렉션을 삭제하고자 할 때
  (*만약, db안에 존재하고 있는 컬렉션이 아무것도 없어질 때, db는 자동으로 삭제)
- db.컬렉션명.isCapped() : 해당 컬렉션의 capped 여부확인가능!!!
- db.컬렉션명.renameCollection("변경하고 싶은 이름") : 컬렉션 이름 변경

#### 5. MongoDB의 주요 테이터 타입
- String : 문자열 데이터 // name : "David"
- Integer : 정수형 데이터 // age : 20
- Boolean : 논리형 데이터 // isStudent : true (or false)
- Dobule : 실수자료형 데이터 (*부동소수점 형태의 값) // rating : 4.5

  (*hobbies : ["reading","music"])
- Object : 객체 데이터 // key & value로 구성되어있는 자료형
  (* address : {city : "Seoul", country : "South Korea"}
- Null : Null 데이터 저장할 수 있는 데이터 형태
