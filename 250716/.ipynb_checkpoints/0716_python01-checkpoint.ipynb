```python
import json
import requests
import pprint    #가독성 좋은 출력을 해줌
import openpyxl

client_id = "pHMhVZYNqEwNhEsbYGPi"
client_secret = "rCEnf8eOmU"

# start = 1
# num = 0

start,num =1,0

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions["B"].width = 100
excel_sheet.column_dimensions["C"].width = 100
excel_sheet.append(["No","제목","링크"])

for index in range(10) :
    start_number = start + (index * 100)      
    naver_open_api = "https://openapi.naver.com/v1/search/shop.json?query=무신사&display=100&start=" + str(start_number)
    header_params = {"X-Naver-Client-Id":client_id,"X-Naver-Client-Secret":client_secret}
    
    res = requests.get(naver_open_api, headers = header_params)
    
    if res.status_code == 200 :
        data = res.json()
        # pprint.pprint(data)
       #  print(data["items"][0]["title"])
        for item in data["items"] :
            num += 1
            excel_sheet.append([num,item["title"],item["link"]])
            print(num, item["title"],item["link"])
    else :
        print("Error Code : ", res.status_code)

excel_file.save("shopping_with_무신사.xlsx")
excel_file.close()
```

```python
import json
import requests

client_id = "pHMhVZYNqEwNhEsbYGPi"
client_secret = "rCEnf8eOmU"

start,num =1,0

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions["B"].width = 80
excel_sheet.column_dimensions["C"].width = 20
excel_sheet.column_dimensions["D"].width = 100
excel_sheet.append(["No","제목","날짜","링크"])

for index in range(10) :
    start_number = start + (index * 100)      
    naver_open_api = "https://openapi.naver.com/v1/search/blog.json?query=홈웨어&display=100&start=" + str(start_number)
    header_params = {"X-Naver-Client-Id":client_id,"X-Naver-Client-Secret":client_secret}
    
    res = requests.get(naver_open_api, headers = header_params)

    if res.status_code == 200 :
        data = res.json()
        
        for item in data["items"] :
            num += 1
            excel_sheet.append([num,item["title"],item["postdate"],item["link"]])
            print(num, item["title"],item["postdate"],item["link"])
           
    else :
        print("Error Code : ", res.status_code)

excel_file.save("reviews_of_홈웨어.xlsx")
excel_file.close()
```
